```python
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

num_rows = 10000

locations = ['Ahmedabad', 'Surat', 'Mumbai', 'Delhi', 'Pune', 'Bangalore', 'Hyderabad', 'Kolkata', 'Chandigarh', 'Jaipur']
colleges = ['Chandigarh University', 'Gujarat Institute Of Technology', 'Nirma University', 'CPET Academy', 'St Xaviers College', 'Indus University']
years_of_study = ['1st', '2nd', '3rd', '4th']
program_interests = ['Data Science', 'Robotics', 'AI', 'Electric Vehicle']
lead_sources = ['Instagram', 'LinkedIn', 'College Collaboration', 'Google Form', 'Mass-Mailing', 'WhatsApp']


data = {
    'Lead ID': [f'LD{i}' for i in range(1, num_rows + 1)],
    'Location': [random.choice(locations) for _ in range(num_rows)],
    'College': [random.choice(colleges) for _ in range(num_rows)],
    'Year of Study': [random.choice(years_of_study) for _ in range(num_rows)],
    'Program Interest': [random.choice(program_interests) for _ in range(num_rows)],
    'Lead Source': [random.choice(lead_sources) for _ in range(num_rows)],
    }

df = pd.DataFrame(data)

df.to_csv('lead_dataset.csv', index=False)
print("Dataset created and saved as 'lead_dataset.csv'")

```

```python
location_lead_source = df.groupby(['Location', 'Lead Source']).size().unstack().fillna(0)

college_lead_source = df.groupby(['College', 'Lead Source']).size().unstack().fillna(0)


year_of_study_lead_source = df.groupby(['Year of Study', 'Lead Source']).size().unstack().fillna(0)

print("\nLead Source Trends by Location:\n", location_lead_source)
print("\nLead Source Trends by College:\n", college_lead_source)
print("\nLead Source Trends by Year of Study:\n", year_of_study_lead_source)

plt.figure(figsize=(12, 8))
sns.heatmap(location_lead_source, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Lead Source Trends by Location')
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(college_lead_source, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Lead Source Trends by College')
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(year_of_study_lead_source, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Lead Source Trends by Year of Study')
plt.show()
```

```python
np.random.seed(42)
df['Converted'] = np.random.choice([0, 1], size=len(df), p=[0.7, 0.3])

conversion_rates = df.groupby('Program Interest')['Converted'].mean()

new_leads = {program: 1000 for program in df['Program Interest'].unique()}

lead_acquisition_cost = {
    'Data Science': 20,
    'AI': 25,
    'Robotics': 15,
    'Electric Vehicle': 30
                }


program_importance = {
    'Data Science': 8,
    'AI': 7,
    'Robotics': 6,
    'Electric Vehicle': 9
    }


projected_conversions = {
    program: new_leads[program] * conversion_rates[program]
    for program in new_leads
    }


lead_acquisition_cost_total = {
    program: new_leads[program] * lead_acquisition_cost[program]
    for program in new_leads
    }


total_budget = 100000


total_projected_conversions = sum(projected_conversions.values())
program_weights = {
program: (projected_conversions[program] * program_importance[program]) / total_projected_conversions
                    for program in projected_conversions
                    }

budget_allocation = {
    program: total_budget * program_weights[program]
    for program in program_weights
    }


print("Conversion Rates for Each Program:\n", conversion_rates)
print("\nProjected Conversions for Each Program:\n", projected_conversions)
print("\nTotal Lead Acquisition Cost for Each Program:\n", lead_acquisition_cost_total)
print("\nBudget Allocation for Each Program:\n", budget_allocation)
```
